```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
from scipy.fft import fft, ifft, fft2, fftn, ifft2
```

```python

images = ["box1.png", "box2.png", "box3.png", "box4.png", "oval1.png", "oval2.png", "oval3.png", "rectangle1.png", "rectangle2.png","rounded1.png", "triangle1.png"]
for img in images:
    fig = plt.figure()
    ax = plt.axes()
    
#     ax.set_title(img, loc='left')
    image = cv2.imread(img)
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    
    transform = []
    
    plt.title(img)
    plt.imshow(np.array(gray, dtype=float), cmap='gray', vmin=0, vmax=255.0)
    plt.show()
    for row in gray:
        transform.append(fft(row))
    transform = fft2(gray)
    
    plt.title(img + " Transformed")
#     plt.imshow(np.array(transform, dtype=float))
    plt.imshow(np.array(transform, dtype=float), cmap='gray', vmin=0, vmax=255.0)
    plt.show()
    
```
